# Simulating Random Walks with NumPy

A random walk is a stochastic process that describes a path consisting of a succession of random steps. This notebook will simulate random walks, visualize a single path, simulate multiple walks in parallel, and analyze the distribution of final positions to confirm the Central Limit Theorem.

## Import Required Libraries

Import NumPy for numerical calculations and Matplotlib for plotting the walk.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

## Simulate a Single Random Walk

Generate a sequence of random steps and compute the cumulative sum to simulate the path of a random walk.

```python
# Number of steps in the random walk
n_steps = 1000

# Generate random steps: +1 or -1 with equal probability
steps = np.random.choice([1, -1], size=n_steps)

# Compute the cumulative sum to get the position at each step
position = np.cumsum(steps)

print(f"Position after {n_steps} steps: {position[-1]}")
```

## Visualize the Walk

Plot the path of the random walk over time using Matplotlib to see the trajectory.

```python
plt.figure(figsize=(12, 5))
plt.plot(position, linewidth=1.5)
plt.title("1D Random Walk Path", fontsize=14)
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(alpha=0.3)
plt.show()
```

## Simulate Multiple Walks

Use NumPy's vectorized operations to efficiently simulate a large number of random walks at once.

```python
# Number of walkers and steps per walk
n_walkers = 50
n_steps = 500

# Generate steps for all walkers: shape (n_walkers, n_steps)
all_steps = np.random.choice([1, -1], size=(n_walkers, n_steps))

# Compute cumulative sum along the time axis (axis=1)
all_positions = np.cumsum(all_steps, axis=1)

print(f"Shape of positions array: {all_positions.shape}")
print(f"Final positions of first 5 walkers: {all_positions[:, -1][:5]}")
```

## Visualize Multiple Walks

Plot all walker trajectories on a single chart to observe the diffusion process.

```python
plt.figure(figsize=(12, 6))

# Plot each walker's trajectory with low opacity
for i in range(n_walkers):
    plt.plot(all_positions[i], linewidth=0.8, alpha=0.4)

plt.title(f"Ensemble of {n_walkers} Independent Random Walks", fontsize=14)
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(alpha=0.3)
plt.show()
```

## Analyze the Distribution

Plot a histogram of the final positions of the walks and compare it to the expected normal distribution.

```python
# Use more walkers for a better histogram
n_walkers = 5000
n_steps = 500

all_steps = np.random.choice([1, -1], size=(n_walkers, n_steps))
final_positions = np.sum(all_steps, axis=1)

plt.figure(figsize=(10, 6))
count, bins, patches = plt.hist(final_positions, bins=40, density=True, alpha=0.7, edgecolor="black")

# Overlay expected normal distribution
x = np.linspace(final_positions.min(), final_positions.max(), 200)
# For a random walk of N steps, variance = N (since each step has variance 1)
expected = np.exp(-(x**2) / (2 * n_steps)) / np.sqrt(2 * np.pi * n_steps)
plt.plot(x, expected, "r-", linewidth=2, label="Normal distribution")

plt.title(f"Distribution of Final Positions ({n_walkers} walks, {n_steps} steps)", fontsize=14)
plt.xlabel("Final Position")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
